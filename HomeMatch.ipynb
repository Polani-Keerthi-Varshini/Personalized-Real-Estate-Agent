### Creating a Vector Database and Storing Listings

The project must demonstrate the creation of a vector database and successfully storing real estate listing embeddings within it. The database should effectively store and organize the embeddings generated from the LLM-created listings.

```python
# create dataframe from listings to save in vector databse
listing_df = pd.DataFrame(listings,index =range(len(listings)))
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer
from typing import Union

MODEL_NAME = 'paraphrase-MiniLM-L6-v2'

model = SentenceTransformer(MODEL_NAME,device="cpu")

def generate_embeddings(input_data: Union[str, list[str]]) -> np.ndarray: 
    
    '''
    Description - generate embeddings from input data using sentencetransfomer model
    '''
    
    embeddings = model.encode(input_data)
    return embeddings
```

```python
# save embeddings in a list using descriptions of peopwert and neighbourhood togetger combining both
embeddings = generate_embeddings(listing_df['Description']+listing_df['NeighborhoodDescription'])
embeddings = [embed for embed in embeddings]
```

```python
# add embedding list to the datafram
listing_df['embeddings'] = embeddings
```

```python
# connect to lancedb databse
db = lancedb.connect("~/.lancedb")

# define data structure of db
class Property(LanceModel):
    Neighborhood:str
    Price:str
    Bedrooms:int
    Bathrooms:float
    HouseSize:str
    Description: str
    NeighborhoodDescription:str
    embeddings:Vector(384)
        
```

```python
table_name = "Property"
db.drop_table(table_name, ignore_missing=True)
# create new table using schema and data 
table = db.create_table(table_name,schema=Property.to_arrow_schema(),data=listing_df)
```

### Semantic Search of Listings Based on Buyer Preferences

The application must include a functionality where listings are semantically searched based on given buyer preferences. The search should return listings that closely match the input preferences.


```python
def retrieve_data(answers):
    '''
    Description: retrieve the related listing from database and use it in chat chain
    Args:
        answers- string taken from humen question and answers and use it for finding similarty
    '''
    buyer_preferences = ' '.join(answers)
    query = model.encode(buyer_preferences)
    matched_listing = table.search(query).limit(1).to_pandas()
    
    values = zip(*[matched_listing["Neighborhood"],
                    matched_listing["Price"], 
                    matched_listing["Bedrooms"], 
                    matched_listing["Bathrooms"], 
                    matched_listing["HouseSize"],
                    matched_listing["Description"],
                    matched_listing["NeighborhoodDescription"]])

    keys = Property.model_fields.keys()
    data = [dict(zip(keys, v)) for v in values][0]
    
    return data


questions = [   
                "How big do you want your house to be?" 
                "What are 3 most important things for you in choosing this property?", 
                "Which amenities would you like?", 
                "Which transportation options are important to you?",
                "How urban do you want your neighborhood to be?",   
            ]
answers = [
    "A comfortable three-bedroom house with a spacious kitchen and a cozy living room.",
    "A quiet neighborhood, good local schools, and convenient shopping options.",
    "A backyard for gardening, a two-car garage, and a modern, energy-efficient heating system.",
    "Easy access to a reliable bus line, proximity to a major highway, and bike-friendly roads.",
    "A balance between suburban tranquility and access to urban amenities like restaurants and theaters."
    ]
```

### Use of LLM for Generating Personalized Descriptions

The submission must utilize an LLM to generate personalized descriptions for the real estate listings based on buyer preferences. The descriptions should be unique, appealing, and tailored to the preferences provided.

```python
# save chat history 
history = ChatMessageHistory()
history.add_user_message(f"""You are AI that will recommend houses as a real state agent based on question answering from user. Ask user {len(questions)} questions""")
for i in range(len(questions)):
    history.add_ai_message(questions[i])
    history.add_user_message(answers[i])

# define conversation summary
summary_memory = ConversationSummaryMemory(
    llm=llm,
    memory_key="preference_summary", 
    input_key="input",
    buffer=f"The human answered {len(questions)} personal questions). Use them to provide answer based on the most recommened house from listings and tailor you response based on the facts from user answers.",
    return_messages=True)

# define Mementobuffersummary
class MementoBufferMemory(ConversationBufferMemory):
    def save_context(self, inputs: Dict[str, Any], outputs: Dict[str, str]) -> None:
        input_str, output_str = self._get_input_output(inputs, outputs)
        self.chat_memory.add_ai_message(output_str)
    
conversational_memory = MementoBufferMemory(
    chat_memory=history,
    memory_key="questions_and_answers", 
    input_key="input"
)

# Combined memories
memory = CombinedMemory(memories=[conversational_memory, summary_memory])

# 
RECOMMENDER_TEMPLATE = """The following is a friendly conversation between a human and an AI Real State Agent. 
                        The AI is follows human instructions and provides answers in a template predefined based on the recommendation 
                        from the house listings from database and creates answer in the same structure as listing template
                        with personalizing the user preferences,
                        subtly emphasizing aspects of the property that align with what the buyer is looking for

Summary of preferences:
{preference_summary}
Personal Questions and Answers:
{questions_and_answers}
Human: {input}
AI:"""
PROMPT = PromptTemplate(
    input_variables=["preference_summary", "input", "questions_and_answers"],
    template=RECOMMENDER_TEMPLATE
)
recommender = ConversationChain(llm=llm, verbose=True, memory=memory, prompt=PROMPT)

matched_listing = retrieve_data(answers)

listing_match = """
=================================
Recommended from listing database according to this template:

'':{Neighborhood}
 Price:{Price}
 Bedrooms: {Bedrooms}
 Bathrooms: {Bathrooms}
 HouseSize : {HouseSize}
 Description:{Description}: 
 NeighborhoodDescription:{NeighborhoodDescription}

""".format(**data)
prediction = recommender.predict(input=listing_match)
print(prediction)
    
```
